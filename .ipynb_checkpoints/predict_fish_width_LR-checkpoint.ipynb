```python
# importing some neccesary libraries, opening csv file and checking it

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# make a variable of LinearRegression

LR = LinearRegression()

# make a variable of StandardScaler()

scaler = StandardScaler()

# opening my csv file

data = pd.read_csv('Fish_market.csv') # opening my csv file

# print(data.isna().any()) 

# checking my data is it clean or not

# define what is my feature and gt (ground truth) or target value in the given dataset

features = ['Weight', 'Length1', 'Length2', 'Length3', 'Height']
gt = ['Width']

X, y = data[features], data[gt] # modifying my values into new one

# split my values into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 9)

# check the numbers of values after spliting it.

print(f'Check the numbers of values after spliting it. \n{45* "-"}')

print (f'X train set has {X_train.shape} values. ')
print (f'X test set has {len(X_test)} values. \n ')

print (f'y train set has {y_train.shape} values. ')
print (f'y test set has {len(y_test)} values. \n\n')

scaler.fit_transform(X_train, y_train) # the dataset has different values so encode them by using StandardScaler

LR.fit(X_train, y_train) # applying LinearRegression

y_pred = LR.predict(X_test) # predict y values by using x test values

# checking accurancy of my LinearRegression
print(f'Checking accurancy of LinearRegression. \n{45* "-"}')

R2 = r2_score(y_test, y_pred)# how my data can fit to my data
print(f'The R2 value: { R2 }') 

MSE = mean_squared_error(y_test, y_pred) # avarage squared distance between actual value and pred. value actua MSE = GT - Pred
print(f'The MeanSquaredError value: { MSE }') 

MAE = mean_absolute_error(y_test, y_pred) # avarage absolute distance between actual value and pred.
print(f'The MeanAbsoluteError value: { MAE }') 

INTERCEPT = LR.intercept_ # starting point of RL in Y-AXIS
print(f'The INTERCEPT value: { INTERCEPT }\n{45* "-"}') 

# illustrate with the graph 

plt.scatter(y_test, y_pred, alpha = 0.8)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color = 'black')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('The predicted/The actual values of Fish Width by using Linear Regression')
plt.show()

```
